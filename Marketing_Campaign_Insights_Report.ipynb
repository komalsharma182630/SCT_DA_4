```python
# MARKETING FUNNEL
funnel_stages = {
    'Website Visitors' : df['NumWebVisitsMonth'].sum(),
    'Web Purchases'    : df['NumWebPurchases'].sum(),
    'Catalog Purchases': df['NumCatalogPurchases'].sum(),
    'Store Purchases'  : df['NumStorePurchases'].sum(),
    'Campaign Response': df['Response'].sum()
}

stages = list(funnel_stages.keys())
values = list(funnel_stages.values())

plt.figure(figsize=(8, 5))
bars = plt.barh(stages[::-1], values[::-1], color=['#4C72B0','#55A868','#C44E52','#8172B2','#CCB974'])
plt.xlabel('Count')
plt.title('Marketing Funnel — Visitors to Conversions')
for bar, val in zip(bars, values[::-1]):
    plt.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2,
             f'{val:,}', va='center', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# ROI / EFFECTIVENESS PER CHANNEL

# Average spend of buyers per channel
web_avg     = df[df['NumWebPurchases']     > 0]['Total_Spending'].mean()
catalog_avg = df[df['NumCatalogPurchases'] > 0]['Total_Spending'].mean()
store_avg   = df[df['NumStorePurchases']   > 0]['Total_Spending'].mean()

# Conversion rate (used channel AND responded to campaign)
web_conv     = (df[(df['NumWebPurchases']     > 0) & (df['Response'] == 1)].shape[0]
                / df[df['NumWebPurchases']     > 0].shape[0]) * 100
catalog_conv = (df[(df['NumCatalogPurchases'] > 0) & (df['Response'] == 1)].shape[0]
                / df[df['NumCatalogPurchases'] > 0].shape[0]) * 100
store_conv   = (df[(df['NumStorePurchases']   > 0) & (df['Response'] == 1)].shape[0]
                / df[df['NumStorePurchases']   > 0].shape[0]) * 100

# Build summary table
roi_df = pd.DataFrame({
    'Channel'          : ['Web', 'Catalog', 'Store'],
    'Avg_Spend (₹)'   : [round(web_avg,2), round(catalog_avg,2), round(store_avg,2)],
    'Conversion_Rate%' : [round(web_conv,2), round(catalog_conv,2), round(store_conv,2)]
})

# ROI Score = Avg Spend × Conversion Rate (proxy for value per channel)
roi_df['ROI_Score'] = (roi_df['Avg_Spend (₹)'] * roi_df['Conversion_Rate%'] / 100).round(2)
roi_df = roi_df.sort_values('ROI_Score', ascending=False).reset_index(drop=True)
print(roi_df)
```

```python
# ROI SCORE BAR CHART
plt.figure(figsize=(7, 4))
colors = ['gold' if i == 0 else 'steelblue' for i in range(len(roi_df))]
plt.bar(roi_df['Channel'], roi_df['ROI_Score'], color=colors, edgecolor='black')
plt.title('ROI Score by Marketing Channel\n(Avg Spend × Conversion Rate)', fontsize=13)
plt.ylabel('ROI Score')
plt.xlabel('Channel')
for i, val in enumerate(roi_df['ROI_Score']):
    plt.text(i, val + 1, str(val), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
print("\n✅ Highest ROI Channel:", roi_df.iloc[0]['Channel'])
```
